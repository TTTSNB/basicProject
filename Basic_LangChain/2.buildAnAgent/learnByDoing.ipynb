***CREATING A PANDAS AGENT TO ANALYZE DATA*** 

**Step 1: Setting up your API Key**

```python
from dotenv import load_dotenv, find_dotenv # for finding and loading environment variables
import pandas as pd # for data manipulation

load_dotenv(find_dotenv()) # load environment variables from .env file
```

**Step 2:** Import the necessary libraries for our agent.

```python
from langchain.chat_models import ChatOpenAI # for chatbot models
from langchain.agents import create_pandas_dataframe_agent # for creating agents from pandas dataframes
```

**Step 3:** Create a pandas dataframe and import your data.csv data file (you'll have to provide your own).

```python
df=pd.read_csv('data.csv') # load data from csv file
```

**Step 4:** Display the dataframe to make sure it looks right.

```python
df
```

**Step 5:** Instantiate the chat model.

```python
chat = ChatOpenAI( # create chatbot model
    model_name="gpt-4", # specify model name
    temperature=0.0 # specify model temperature
)
agent = create_pandas_dataframe_agent(chat, df, verbose=True) # create agent from dataframe and chatbot model
```

**Step 6:** Ask a question to the chat model.

```python
agent.run("What is the total revenue generated by the company?")
```
